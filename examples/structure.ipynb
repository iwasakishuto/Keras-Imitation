## Data

```python
from kerasy.utils import generateSeq
```

```python
seq = generateSeq(size=150, nucleic_acid="RNA", seed=0)
```

## Nussinov

```python
from kerasy.Bio.structure import Nussinov
```

```python
model = Nussinov()
```

```python
model.load_params()
```

```python
model.params()
```

### Predict secondary structure.

```python
model.predict(seq)
```

```python
Z = model.ConstrainedMaximize(seq)
```
