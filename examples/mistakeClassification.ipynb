Trains a simple neural net on the Titanic dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from kerasy.datasets import mnist
from kerasy.models import Sequential
from kerasy.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Input
from kerasy.utils import CategoricalEncoder
```

```python
from kerasy.utils import generateWholeCakes
```

```python
K=5
N=200
seed=0
```

```python
x_train, y_train = generateWholeCakes(cls=K, N=N, seed=seed, plot=True, figsize=(4,4))
```

```python
batch_size = 32
epochs = 20
```

```python
model = Sequential()
model.add(Input(input_shape=(2,)))
model.add(Dense(8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dropout(keep_prob = 0.5))
model.add(Dense(8, activation='relu'))
model.add(Dense(K, activation='softmax'))
```

```python
model.compile(optimizer='sgd', loss="categorical_crossentropy")
```

```python
model.summary()
```

```python
model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,)
```

```python
np.argmax(model.predict(x_train), axis=1)
```
