## Data

```python
import numpy as np
from kerasy.utils import generateSeq
```

```python
n = 150
n_same = 20
seed = 1
```

```python
seqX, seqY = generateSeq(size=(2,n), nucleic_acid="RNA", seed=seed)
```

## Needleman-Wunsch-Gotoh

```python
from kerasy.Bio.alignment import NeedlemanWunshGotoh
```

```python
model = NeedlemanWunshGotoh()
```

```python
model.load_params()
```

```python
model.params()
```

```python
model.align(seqX, seqY)
```

## Backward-Needleman-Wunsch-Gotoh

```python
from kerasy.Bio.alignment import BackwardNeedlemanWunshGotoh
```

```python
model = BackwardNeedlemanWunshGotoh()
```

```python
model.load_params()
```

```python
model.params()
```

```python
model.align(seqX, seqY)
```

## Smith-Waterman-Gotoh

```python
from kerasy.Bio.alignment import SmithWaterman
```

## Pair HMM

```python
from kerasy.Bio.alignment import PairHMM
```

```python
model = PairHMM()
```

```python
model.load_params()
```

```python
model.params()
```

```python
model.align(seqX, seqY)
```
